```python
 import pandas as pd
from sklearn.datasets import load_iris #datasets is subset and sklearn is the library 
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns = data.feature_names)
df["target"] = data.target
df.head()
```

visualizing basic features to understand the data. 

```python
#scatter plot of sepal length and width

import matplotlib.pyplot as plt
plt.scatter(df["sepal length (cm)"], df ["sepal width (cm)"], 
            c=data.target)

#c = data.target creates different colors by using parameter "c" and the numbers target labelled 0,1 and 2 for each species
```
